```python
import numpy as np
import pandas as pd
import folium

from folium.plugins import HeatMap
from matplotlib import pyplot as plt
from sklearn.metrics.pairwise import haversine_distances as haversine
```

# Objetivo
O objetivo principal deste notebook é realizar a predição do custo de uma corrida de taxi em Nova York utilizando como base os dados fonecidos pelo Kaggle.

# Obtendo os dados de treino
Os dados de treinos são tão extensos que não foi possível ler todos. 16 GB não são suficientes

```python
trainDf = pd.read_csv('../../datasets/new-york-city-taxi-fare-prediction/train.csv', nrows=50000)
trainDf.head()
```

Estamos lendo apenas uma amostra dos dados para apenas dar podermos observar como eles estão estruturados.

### Podemos observar que exiistem as seguintes features:
 - Key: Concatenação dos atributos data e hora com adição de um "id", pois as vigens poderiam ser iniciadas no mesmo momento.
 - pickup_datetime: Momento no qual o(s) passageiro(s) iniciaram a viagem
 - pickup_longtude: longitude de onde os passgeiros foram pegos
 - pickup_latitude: latitude de onde os passgeiros foram pegos
 - dropoff_longtude: longitude de onde os passgeiros foram deixados
 - dropoff_latitude: latitude de onde os passgeiros foram deixados
 - passenger_count: Quantidade de passageiros na viagem 

# Preprocessamento dos dados
Transformaremos a coluna pickup_datetime em duas, separando os valores em dia e o momento do dia.

A motivação de está fazendo isso é por que acredito que influencie os preços de acordo com o horário.

```python
datetime = [trainDf['pickup_datetime'].apply(lambda x: np.array(x.split(' ')[:2]))]
datetime = np.array(datetime)[0]
trainDf['date'] = datetime[:, 0]
trainDf['time'] = datetime[:, 1]
trainDf.drop('pickup_datetime', axis=1, inplace=True)
trainDf.drop('key', axis=1, inplace=True)
```

```python
trainDf.head()
```

# Apresentação de gráficos.
### Mapas
Para iniciar a apresentação dos mapas primeiramente tiraremos o ponto médio das coordenadas para centralizar a visão do mapa

```python
# a formação de uma cordenada é feita por por uma latitude e uma longitude -> coord(latitude, longitude)
coordenadaMedia = (40.750424, -73.983253) #Coordenadas fornecidas pelo google
coordenadaMedia 
```

```python
mapaDeCalor = folium.Map(coordenadaMedia, zoom_start=11, width='100%', height='100%')
mapaDeCalor.add_child(HeatMap(trainDf[['pickup_latitude', 'pickup_longitude']]
                              .apply(tuple, axis=1), radius=10))
mapaDeCalor 
# Observação: só consegui mostrar o mapa para apenas 50.000 pontos, o folium não suporta muito mais que isso
```

Com este mapa podemos observar que a maioria das viagens começam em três pontos principais 
 - No centro de Manhattan
 - No aeroporto de LaGuardia
 - No Aeroporto internacional John F. Kennedy

```python
mapaDeCalor = folium.Map(coordenadaMedia, zoom_start=11, width='100%', height='100%')
mapaDeCalor.add_child(HeatMap(trainDf[['dropoff_latitude', 'dropoff_longitude']]
                              .apply(tuple, axis=1), radius=10))
mapaDeCalor 
```

Com este outro mapa é possível observar que as viagens terminam em pontos muito mais distribuídos do que no mapa anterior, mas a concentração máxima de pontos ainda é no centro de Manhattan

### Gráficos
Grafico de relação entre preço e hora, será utilizado a média do preço para cada hora.

```python
trainDf['hour'] = [trainDf['time'].apply(lambda x: int(np.array(x.split(':')[0])))][0]
rel = trainDf.groupby(['hour'])['fare_amount'].agg(lambda x: x.unique().sum()/x.nunique())
```

```python
plt.plot(rel, 'bo-');
```

Como podemos observar, a média de valores mais altas ficam por volta do meio dia. E isso não possui relevacia nenhuma

```python
trainDf['dist_vet'] = trainDf.apply(lambda x: haversine([[x['pickup_latitude'],x['pickup_longitude']],[x['dropoff_latitude'],x['dropoff_longitude']]])[0][1] * 6371000, axis=1)
trainDf['dist_vet']
```
